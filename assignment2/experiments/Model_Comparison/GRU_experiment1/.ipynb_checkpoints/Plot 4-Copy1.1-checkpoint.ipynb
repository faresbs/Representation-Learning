```python
import numpy as np
```

```python
Model = "RNN+GRU's"
```

```python
from matplotlib import pyplot as plt
def learning_curves(x):
    # Training & validation accuracy values
    plt.plot(x['Train PPL'])
    plt.plot(x['Val PPL'])
    plt.title(Model+' PPL per epoch')
    plt.ylabel('PPL')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
    
    plt.plot(x['Wall-clock'],x['Train PPL'])
    plt.plot(x['Wall-clock'],x['Val PPL'])
    plt.title(Model+' PPL per wall-clock-time')
    plt.ylabel('Perplexity')
    plt.xlabel('Wall-clock-time (seconds)')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
```

```python
import csv
with open('log2_GRU.csv') as f:
    reader = csv.reader(f)
    next(reader) # skip header
    data = []
    for row in reader:
        data.append(row)
```

```python
import pandas as pd 
data = pd.read_csv('log2.csv', sep=';')
```

```python
print(data[0:10])
```

```python
learning_curves(data)
```
